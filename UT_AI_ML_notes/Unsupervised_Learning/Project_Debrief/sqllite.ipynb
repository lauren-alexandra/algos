```python
query = '''
select b.customertype, a.seat_comfort, count(*)
from survey a InNer join flight b on a.id = b.id
group by 1, 2
order by 1, 2
'''
pd.read_sql_query(query,
                 engine)
```

```python
query = '''
select a.*
from survey a inner join flight b on a.id = b.id
where b.CustomerType = 'Loyal Customer'
'''
df2 = pd.read_sql_query(query,
                 engine)
```
